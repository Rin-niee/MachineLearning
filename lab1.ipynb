##**4. Визуализация распределения переменных**##

```python
sns.histplot(df['AGE_50'], kde=True)
plt.title('Распределение AGE_50')
plt.xlabel('AGE_50')
plt.ylabel('Частота')
plt.show()
```

**2. Столбчатая диаграмма для категориальной переменной DEATH**

```python
sns.countplot(x='DEATH', data=df)
plt.title('Распределение DEATH')
plt.xlabel('Death')
plt.ylabel('Количество')
plt.show()
```

**Боксплот для числовой переменной `SBP_50` :**

```python
sns.boxplot(x=df['SBP_50'])
plt.title('Boxplot SBP_50')
plt.xlabel('SBP_50')
plt.show()
```

**Наблюдения:**

Распределение считается околонормальным, так как в пределах 140 есть небольшое отклонение по  графику. Выбросов совершенно немного, медиана сосредоточена на 120, между 100 и 140

**Корреляционную матрицу для всех числовых переменных и визуализация её с помощью тепловой карты:**

```python
# Корреляционная матрица
correlation_matrix = df.corr().round(1)
# Визуализация тепловой карты
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Корреляционная матрица')
plt.show()
```
